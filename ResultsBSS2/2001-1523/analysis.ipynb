```python
def VLSBtobio(LSB):
    Vhw=0.6/400*(LSB-700)+1
    Vb=(Vhw-1.5)/15
    return Vb
```

```python
def getfreq(data,thres):
    data2=data-thres
    dif=data2[1:]*data2[:-1]
    return np.where(dif<0)[0].size/2
```

```python
Lfreq=[getfreq(LL[i,5],400) for i in range(9)]
```

```python
plt.scatter(1,30)
for j in range (9):
    plt.plot(cur,[getfreq(LL[i,j],400) for i in range(9)],'--.',label=str(cond[j]))
plt.legend()
plt.xlabel("Biological current (nA)")
plt.ylabel('Firing rate (Hz)')
```
