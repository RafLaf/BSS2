```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Lv=list()
for i in range (1,12):
    df = pd.read_csv(str(i)+".dat",sep='\t',header=2)
    data=np.array(df)
    Lv.append(data)
    plt.plot(data[:,0],data[:,1],label=str(i))
plt.legend()
```

```python
Lim=list()
for i in range (1,12):
    df = pd.read_csv(str(i)+"b.dat",sep='\t',header=2)
    data=np.array(df)
    Lim.append(data)
    plt.plot(data[:,0],data[:,1],label=str(i))
plt.legend()
```

```python
def detect(x,thresh):
    '''
    returns the number a threshold is passed increasing or decreasing
    '''
    x-=thresh
    a=x[1:]*x[:-1]
    return np.where(a<0)[0].size
```

```python
np.array(Lv).shape
```

```python
Lthres=list()
for x in Lv:
    y=x[:,1]
    Lthres.append(detect(y,0))
```

```python
Lthres
```

```python
plt.plot([0.75,0.1,0.3,0.5,0.7,0.9,1.1,1.3,1.5,1.7,1.9],np.array(Lthres)/2,'.')   
plt.xlabel('input current (nA)')
plt.ylabel('nb of spike ')
```
