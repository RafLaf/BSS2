```python
def detect(x,thresh):
    '''
    returns the number a threshold is passed increasing or decreasing
    '''
    x2=x-thresh
    a=x2[1:]*x2[:-1]
    return np.where(a<0)[0].size
```

```python
nspikestep,nspikeramp=[],[]
for x in stepresponse:
    nspikestep.append(detect(x,20)/2/0.8)
for x in rampreponse:
    nspikeramp.append(detect(x,20)/2/1)    
```

```python
plt.plot(steps,nspikestep,'.')
plt.ylabel('Firing rate (Hz)')
plt.xlabel('Step amplitude (nA)')
np.save("nspikestep.npy",nspikestep)
```

```python
def timebtwSpike(x,thresh=20,dt=0.1):
    x2=x-thresh
    a=x2[1:]*x2[:-1]
    a=np.where(a<0)[0][slice(0,-1,2)]   #let us only consider "increasing threshold crossings"
    dif=dt*(a[1:]-a[:-1])
    return dif
```

```python
for i in range (10,50,10):    
    plt.plot(timebtwSpike(rampreponse[i]),label='ramp end='+str(np.int(ramps[i]*100)/100)+'nA')
plt.xlabel('rank of spike')
plt.ylabel('Time delay between spikes (ms)')
plt.legend()
```

```python
for i in range (10,50,10):    
    plt.plot(timebtwSpike(stepresponse[i]),label='step amp='+str(np.int(ramps[i]*100)/100)+'nA')
plt.xlabel('rank of spike')
plt.ylabel('Time delay between spikes (ms)')
plt.legend()
```
